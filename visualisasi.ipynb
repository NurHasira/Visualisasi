```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Buat data baru
data_baru = {
    "pekerjaan": ["Guru", "Dokter", "Insinyur", "Guru", "Dokter", "Insinyur", "Guru", "Dokter", "Insinyur", "Guru"],
    "gaji_bulanan": [5.5, 12.0, 9.0, 6.0, 11.5, 9.5, 5.8, 12.2, 8.8, 6.2]  # dalam juta rupiah
}
df_pekerjaan = pd.DataFrame(data_baru)

# Visualisasi boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_pekerjaan["pekerjaan"], y=df_pekerjaan["gaji_bulanan"])
plt.title("Boxplot Gaji Bulanan per Jenis Pekerjaan")
plt.xlabel("Jenis Pekerjaan")
plt.ylabel("Gaji Bulanan (juta rupiah)")
plt.grid(True)
plt.show()
```
